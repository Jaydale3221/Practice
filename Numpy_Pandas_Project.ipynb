```python
df = df.rename(columns={"math score": "Math", "reading score": "Reading", "writing score": "Writing"})
```

```python
df.head()
```

```python
df = df.rename(columns = {"Math" : "Maths"})
```

```python
df.head()
```

```python
# Mean, median, max, min
df[["Maths", "Reading", "Writing"]].mean()
df[["Maths", "Reading", "Writing"]].median()
df[["Maths", "Reading", "Writing"]].std()
```

```python
scores = df[["Maths", "Reading", "Writing"]].to_numpy()
np.mean(scores, axis=0)
np.std(scores, axis=0)
np.min(scores, axis=0)
np.max(scores, axis=0)


```

```python
scores
```

```python
# Filter students who scored > 90 in Math

# df["Maths"] > 90
df[df["Maths"] > 90]
```

```python
# Filter students with Reading > 80 AND Writing > 80
df[(df["Reading"] > 80) & (df["Writing"] >80)]
```

```python
# .loc and .iloc
# df.loc[0:8, ["Reading"]]
df.iloc[1:8, 2:10]
```

```python
math_scores = df["Math"].to_numpy()
math_scores[math_scores > 90]
```

```python
# Add Total and Average score columns
df["Total"] = df["Math"] + df["Reading"] + df["Writing"]
df["Average"] = df["Total"] / 3
```

```python
df['Average']
```

```python
# Categorize Performance
df["Performance"] = pd.cut(df["Average"],
                           bins=[0, 60, 80, 100],
                           labels=["Low", "Medium", "High"])
```

```python
df['Performance']
```

```python
# Average Math score by gender
# df.groupby("gender")["Math"].mean()
df.groupby("gender")["Math"].mean()
```

```python
# Multiple aggregations
df.groupby("gender")[["Math","Reading","Writing"]].agg(["mean", "max", "min"])
```
